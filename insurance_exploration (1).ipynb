```python
#need to convert the categorical variables into dummies so that they can be included in the linear regression model
med_dum = pd.get_dummies(med_data)
```

```python
med_dum.columns.values
```

```python
#now lets try it with a fancy package that formats the output and calculates all the statistics of interest automatically
import statsmodels.api as sm
from scipy import stats
```

```python
ind_vars = ['age','bmi','children','sex_female','smoker_no','region_northeast', 'region_northwest', 'region_southeast']
dep_var = 'charges'

x = med_dum[ind_vars]
y = med_dum[dep_var]
X2 = sm.add_constant(x)
est = sm.OLS(y, X2)
est2 = est.fit()
est2.summary()
```
