```python
#need to convert the categorical variables into dummies so that they can be included in the linear regression model
med_dum = pd.get_dummies(med_data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(med_dum, test_size=0.2)
```

```python
med_dum.columns.values
```

```python
ind_vars = ['age','bmi','children','sex_female','smoker_no','region_northeast', 'region_northwest', 'region_southeast']
dep_var = 'charges'
x_train = train[ind_vars]
y_train = train[dep_var]

x_test = test[ind_vars]
y_test = test[dep_var]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)


# The coefficients
coeff = list(zip(ind_vars, lr.coef_ ))

print(pd.DataFrame(coeff, columns = ['Label', 'Coefficient']))
# The mean squared error
print("\nMean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('\nVariance score: %.2f' % r2_score(y_test, y_pred))
```

```python
#a perfect fit would put all the points along the black line so we can see that the model is not very good. The MSE and R2 
#said this already but pictures usually help :) 
plt.plot(y_test, y_pred, 'r.')
plt.plot([0, 40000], [0, 40000], 'k-', lw=2)
plt.show()
```
